```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Using pandas to: Load the data from Excel file , get the information of the Dataframe

```python
df = pd.read_csv(r'D:\StudentsPerformance.csv')
df
```

1.Find the missing value of each column



```python
df_info = df.info()
df_head = df.head()

df_info, df_head
```

2.Find the missing value of each column

```python
missing_values = df.isnull().sum()
missing_values
```

3.Fill the Missing values :

    a.If it is the Number: fill it by 0

```python
numerical_cols = df.select_dtypes(include=['number']).columns
df[numerical_cols] = df[numerical_cols].fillna(0)

df
```

    b.If it is categorical: fill it by none

```python
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna('none')
df
```

```python
cleaned =  df.isna().sum()
cleaned
```

4.Count the number of Male and Female Students

```python
gender_count = df['gender'].value_counts()
gender_count
```

5.Using countplot in seaborn:

    a.Create plot to represent the number of Male and Female student
    


```python
colors = ['#1f77b4', '#ff7f0e']

gender_count = df['gender'].value_counts()
plt.figure(figsize=(8,6))
sns.countplot(data=df, y='gender')
plt.barh(gender_count.index, gender_count.values, color=colors)

plt.title('Number of Male and Female Students')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

    b.Create plot to divide the data by race/ethnicity

```python
custom_palette = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='race/ethnicity', hue='race/ethnicity', palette=custom_palette, legend=False)
plt.title('Distribution by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Count')
plt.show()
```

    c.Create plot to see the distribution of parental education levels split by degree type

```python
custom_palette = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']  # Adjust as needed

plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='parental level of education', hue='parental level of education', palette=custom_palette, legend=False)
plt.title('Distribution of Parental Education Levels')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Parental Level of Education')
plt.xlabel('Count')
plt.show()
```

6.Using scatter plot in seaborn:

    a.Create a plot to show the Math and Reading scores with the color representing gender. 

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='math score', y='reading score', hue='gender')
plt.title('Math Score vs Reading Score by Gender')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.show()
```

    b.Create a plot to show the Math and Writing scores with the color representing gender. 

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='math score', y='writing score', hue='gender')
plt.title('Math Score vs Writing Score by Gender')
plt.xlabel('Math Score')
plt.ylabel('Writing Score')
plt.show()
```

    c.Create a plot to show the Writing and Reading scores with the color representing gender.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='writing score', y='reading score', hue='gender')
plt.title('Writing Score vs Reading Score by Gender')
plt.xlabel('Writing Score')
plt.ylabel('Reading Score')
plt.show()
```

    d.We know that the “total marks” is the target variable in this specific dataset. Using kdeplot in seaborn to create plot of total mark if each students. 

```python
df['total marks'] = df[['math score', 'reading score', 'writing score']].sum(axis=1) / 3

df_melted = df.melt(value_vars=['total marks', 'math score', 'reading score', 'writing score'], 
                    var_name='Score Type', value_name='Score')

plt.figure(figsize=(10, 6))
sns.kdeplot(data=df_melted, x='Score', hue='Score Type', fill=True, palette='Set1')

plt.title('Score Distribution by Score Type')
plt.xlabel('Score')
plt.ylabel('Density')

plt.show()
```
